```python
path_test="C:\\Users\\DELL\\diff_crop_cnn"
```

```python
Categories=["crop","croton","horse bush","screw pine","silverleaf","variegated screwpine"]
```

```python
batch_count=[]
training=[]
def createTrainingData():
    for cat in Categories:
        path=os.path.join(path_test,cat)
        class_num=Categories.index(cat)
        num=1
        for img in os.listdir(path):
            img_array=cv2.imread(os.path.join(path,img))
            img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB)
            training.append([img_array,class_num])
            num+=1
        batch_count.append(num)    
```

```python
createTrainingData()
```

```python
datagen = ImageDataGenerator(
rotation_range = 40,
shear_range = 0.2,
zoom_range = 0.2,
horizontal_flip = True,
width_shift_range= 0.2)
```

```python
training
```

```python
dir_cat={0:'Crop',1:'Croton',2:'Horsebush',3:'Screwpine',4:'Silverleaf',5:'Variegated screwpine'}
for cat in training:
    
    x=cat[0]
    x=x.reshape((1, ) + x.shape)
    i = 0
    for batch in datagen.flow(x, batch_size = 1,save_to_dir ='dir'+dir_cat[cat[1]],save_prefix ='image', save_format ='jpeg'):
        i += 1
        if i > 5:
            break
```
